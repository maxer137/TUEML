# Q1



In the notes, you have seen that ridge regression corresponds to solving
$$
\hat{\beta}_{\text{ridge}}
=
\arg\min_{\beta\in\mathbb{R}^p}
\left\{
\|y-X\beta\|^2+\lambda\|\beta\|^2
\right\}.
$$
We now consider a probabilistic view of linear regression and show that it leads to the same solution. Define the dataset as $X = \{ x_{1:n}\}$ and $y = \{y_{1:n}\}$, and assume the following model:
$$
y_i = x_i^T\beta + \varepsilon_i,
\qquad
\varepsilon_i \overset{\text{i.i.d.}}{\sim}\mathcal N(0,\sigma^2),
$$
or equivalently,
$$
y_i\mid x_i,\beta \sim \mathcal N(x_i^T\beta,\sigma^2).
$$
Assume also the prior
$$
\beta \sim \mathcal N(0,\sigma_p^2 I_p).
$$
We define the MAP estimator by
$$
\hat\beta_{\text{MAP}}
=
\arg\max_\beta p(\beta\mid X,y).
$$

**Question:** Show that $\hat\beta_{\text{MAP}}$ is equal to the ridge regression solution for a suitable choice of $\lambda$.

You may use the following facts:

- Bayes’ rule:
  $$
  p(\beta\mid X,y)\propto p(y\mid X,\beta)p(\beta)
  $$

- Conditional independence:
  $$
  p(y\mid X,\beta)=\prod_{i=1}^n p(y_i\mid x_i,\beta)
  $$

- Applying $\log$ preserves the maximizer

- Terms that do not depend on $\beta$ may be ignored when optimizing over $\beta$

Derive the corresponding optimization problem explicitly, and state the relationship between $\lambda$, $\sigma^2$, and $\sigma_p^2$.

**Remark**: Your score will be based only on your response in the field " $\lambda:=\ldots$ ". However, to receive credit, you must still upload your derivation.

==**Ans:** $\lambda := \sigma^2/\sigma^2_p$.==

==Solution from the book:==

- <img src="assets/image-20260410153635541.png" alt="image-20260410153635541" style="zoom:50%;" />

$$
\begin{align}
\hat \beta &= \arg \max -\frac12 \left[ \sigma_p^{-2} ||\beta||^2_2 + \sigma^{-2} ||y - X\beta||^2_2 \right] \\
%
&= \arg \min \left[ \sigma_p^{-2} ||\beta||^2_2 + \sigma^{-2} ||y - X\beta||^2_2 \right] \\
%
&= \arg \min \sigma^2 \left[ \sigma_p^{-2} ||\beta||^2_2 + \sigma^{-2} ||y - X\beta||^2_2 \right] \\
%
&= \arg \min  \left[ \frac{\sigma^2}{\sigma_p^{2}} ||\beta||^2_2 + ||y - X\beta||^2_2 \right] \\
\end{align}
$$
